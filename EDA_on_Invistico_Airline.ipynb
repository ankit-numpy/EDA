# •	EDA of total count of Satisfied vs Unsatisfied responses

```

data_crosstab = pd.crosstab(train[label],train['Gender'], colnames=['class'])
#label='satisfaction'
barplot = data_crosstab.plot.bar(rot=0)
data_crosstab = pd.crosstab(train[label],train['Customer Type'], colnames=['class'])
#label='satisfaction'
barplot = data_crosstab.plot.bar(rot=0)
```

```
data_crosstab2 = pd.crosstab(train[label],train['Type of Travel'])
#label='satisfaction'
barplot = data_crosstab2.plot.bar(rot=0)


data_crosstab2 = pd.crosstab(train[label],train['Class'])
#label='satisfaction'
barplot = data_crosstab2.plot.bar(rot=0)
```

```
data_crosstab2 = pd.crosstab(train[label],train['Seat comfort'])
#label='satisfaction'
barplot = data_crosstab2.plot.bar(rot=0)
```

```
train['cat']= pd.cut(train.Age,bins=[0,2,18,60,110],labels=['Toddler','kid','Adult','Senior_c'])
data_crosstab2 = pd.crosstab(train[label],train['cat'])
#label='satisfaction'
barplot = data_crosstab2.plot.bar(rot=0)
```
